## Investigation Overview
In this investigation, I wanted to assess pupils performance score in Mathematics based on their Interest in mathematics - Enjoy Reading, Possession of textbooks,Attribution to failure - material too hard, Gender and Out-of-School Study Time (hours).

Summary of key insights are:
* Those who are in possession of textbooks have a higher mean performance score in Mathematics than those who don't have.
* Most of those who scored higher than 700 in Mathematics opined that a likeness for complex problem solving is "somewhat" - "mostly" like them.
* The lower the degree of likeliness that the reading material is hard, the higher the mean performance score in Mathematics.
* Pupils from OECD countries performed quite better in Mathematics than their counterparts from Non-OECD countries.
* Even though more Females than males agreed to "Interest in Mathematics - Reading", and also more females than males do not perceive the material to be too hard,the male genders have a slightly higher Mathematics performance mean score than the females.
A deeper look at the multivariate relationship shows that, this could be due to the fact that there are more males than females - of those who scored above 700 in mathematics performance.

* Outside of the main variable of interest, I discovered that irrespective of the pupils' level of interest in Mathematics - Enjoy Reading, those who possess textbook are always more than those who don't.



## Dataset Overview
The PISA2012 Data is a survey of students' skills and knowledge as they approach the end of compulsory education. The dataset contains a wealth of information on students' backgrounds and interests, their school and the organisation of education system.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
pisa_df = pd.read_csv("pisa2012.csv",encoding='cp1252')
```

```python
# convert into ordered categorical datatype
def categorize(column_name,column_class):
    """A private function convert the input column of an already defined
     dataframe into a category datatype.
    It's first argument should be the column name,its second argument
    should be a variable that's already assigned an array - this array
    should have been arranged in its right order.
    
    e.g. categorize('cat_var','cat_class')"""
    
    classes = pd.api.types.CategoricalDtype(ordered=True,categories=column_class)
    pisa_df[column_name] = pisa_df[column_name].astype(classes);

# create an array for the order of values
classes_1 = ['Strongly disagree','Disagree','Agree','Strongly agree']
classes_2 = ['Not at all likely','Slightly likely','Likely','Very   Likely']
classes_3 = ['Not at all like me','Not much like me','Somewhat like me','Very much like me','Mostly like me']
# call function to make conversion
categorize('ST29Q01',classes_1)
categorize('ST44Q05',classes_2)
categorize('ST94Q14',classes_3)
```

```python
# data wrangling, removing rows with missing data
pisa_df = pisa_df.dropna(subset=['ST26Q10','ST29Q01','ST44Q05','OUTHOURS','ST94Q14'])
# reset dataframe index
pisa_df = pisa_df.reset_index(drop=True)
```

```python
# data wrangling,removing "Unnamed" column
pisa_df.drop(pisa_df.columns[pisa_df.columns.str.contains("Unnamed")], axis=1, inplace=True)
```

## Distribution of Pupils Performances in Mathematics
Pupils performance score in the dataset takes on a wide range of values from about 20 (at the lowest) to about 960 at the highest. The distribution of Pupils performance score in mathematics takes on a unimodal shape and it's normally distributed.

Students who scored excellently well (above 700) are lower in count, and not much students failed either. Most of their scores fall within 400-600 performance score.

```python
# create the histogram plot using seaborn.displot()
bin_edges = np.arange(pisa_df['PV1MATH'].min(),pisa_df['PV1MATH'].max()+30,30);
sb.displot(pisa_df['PV1MATH'], bins=bin_edges);
plt.xlabel('Pupil Performance in Mathematics (score)');
plt.title('The distribution of Pupils Performances in Mathematics');
plt.show();
```

## Possession of Textbooks vs  Pupil Average Performance in Mathematics
Does possession of textbooks affect the overall performance of pupils in mathematics?

Plotting the relationship between Possession of Texbooks and the mean peformance of pupils shows that: those in Possession of textbooks have a higher (though not too wide a gap) mean performance than those who are not in possession of textbook.

```python
# creating and polishing the barplot
sb.barplot(data = pisa_df, x = 'ST26Q10', y = 'PV1MATH',color=sb.color_palette()[0], errwidth=0);
plt.ylabel('Mean Performance in Mathematics')
plt.xlabel('Possession - Textbooks');
plt.title('Mean Performance Score in Mathematics vs Possession - Textbooks')
ax = plt.gca()
# remove the top and right border lines 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False);
plt.show()
```

## Distribution of Students' Likeness for Solving Complex Problem
The plot shows an increasing trend for the proportion of counts as we move from "Not at all like me" to "Somewhat like", but the trend dropped as we move further to "very much like me" and "mostly like me".
You could be probably thinking what I thought - The distribution is almost normal without been biased to a side.

What if we consider students who had a very high score (above 700), could they really be complex problem solvers?

```python
# obtain the value of each category to calculate proportion
cat_count = pisa_df['ST94Q14'].value_counts(sort=False)
n_pupil = pisa_df['ST94Q14'].value_counts().sum()
sb.countplot(data= pisa_df, y='ST94Q14',color = sb.color_palette()[8])

# a for loop to print the proprtion text on bars
for i in range(cat_count.shape[0]):
    count = cat_count[i]
    pct_string = '{:0.1f}%'.format(100*count/n_pupil)
    plt.text(count+1, i, pct_string, va='center');
    
plt.xticks([])
plt.xlabel('Proportion of Counts');
# remove the top and right border lines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False);
# set label and title
plt.ylabel('Likeliness for Solving Complex Problem')
plt.title("The BarChart Distribution of Pupils for their Likeliness for Solving Complex Problem");
```

## For a performance score of above 700 only: The Distribution of Students' Likeness for Solving Complex Problem
The difference is clear! A higher percentage of the students who scored above 700 opined that "solving complex problem" is somewhat, very much or mostly like them.

```python
# create a subset of the dataframe by Mathematics performance score greater than 700.
pisa_subset = pisa_df[['ST29Q01','PV1MATH','ST94Q14','ST04Q01','ST29Q01']]
pisa_subset_700 = pisa_subset[pisa_subset['PV1MATH']>700]
# obtain the value of each category to calculate proportion
catg_count = pisa_subset_700['ST94Q14'].value_counts(sort=False)
n_pupil = pisa_subset_700['ST94Q14'].value_counts().sum()
sb.countplot(data= pisa_subset_700, y='ST94Q14',color = sb.color_palette()[8])
# a for loop to print the proprtion text on bars
for i in range(catg_count.shape[0]):
    count = catg_count[i]
    pct_string = '{:0.1f}%'.format(100*count/n_pupil)
    plt.text(count+1, i, pct_string, va='center');
    
plt.xticks([])
plt.xlabel('Proportion of Counts');
# remove the top and right border lines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False);
plt.ylabel('Likeliness for Solving Complex Problem')
plt.title("The BarChart Distribution of Pupils who scored above 700 (Mathematics) with their Likeliness for Solving Complex Problem");
```

## Mean Performance vs Attribution to Failure - Material is too Hard
Students who find the material to be hard often lose interest in reading, and when they lose their interest in reading,their performance would most likely drop. The same trend is the case here.
The mean performance decreases with increasing degree of attributing material-is-too-hard to failure.

```python
# Bar plot of Pupil performance against "Failure - material too hard"
sb.barplot(data = pisa_df, x = 'ST44Q05', y ='PV1MATH',color='grey',errwidth=0);
plt.ylabel('Mean Performance in Mathematics (score)')
plt.xlabel('Material too Hard');
plt.title("Mean Performance vs Attribution to Failure - Material is too Hard")
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False);
```

So there we are, as a pupil whose desire is to have a very high score in Mathematics, it is essential to really like solving complex problem, and in the least - try get a possession of textbook. But then,never find the material too hard to study.
